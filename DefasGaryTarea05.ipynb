### Ejercicio 1

```python
import numpy as np
from scipy.optimize import newton

def f(x):
    return x**3 - np.cos(x)

def f_prime(x):
    return 3*x**2 + np.sin(x)

# Método de Newton:
p0_newton = -1
# 1ra Iter: p1
p1_newton = p0_newton - f(p0_newton) / f_prime(p0_newton)
# 2da Iter: p2
p2_newton = p1_newton - f(p1_newton) / f_prime(p1_newton)

# Método de la Secante:
p0_secante = -1
p1_secante = 0
p2_secante = p1_secante - f(p1_secante) * (p1_secante - p0_secante) / (f(p1_secante) - f(p0_secante))


print("Método de Newton:")
print(f"p0 = {p0_newton}")
print(f"p1 = {p1_newton}")
print(f"p2 = {p2_newton}")

print("\nMétodo de la Secante:")
print(f"p0 = {p0_secante}, p1 = {p1_secante}")
print(f"p2 = {p2_secante}")

print("\n¿Se puede usar p0 = 0? \nNo, en absoluto, si es con método Newton, pues daría una división sobre cero")
print("Pero sí se podría usar en el método secante, con la restricción que f(p1) sea distinto de cero")
```

```python
## Parte a
import numpy as np
import scipy.optimize as opt

def f1(x):
  return 3*x - np.exp(x)

def f1prime(x):
  return 3 - np.exp(x)

print("Usando método de Newton:", opt.newton(f1,x0=1.5,fprime=f1prime, tol = 10e-5))
print("Usando método Secante:", opt.newton(f1,x0=1,x1=2,tol = 10e-5))
```

```python
## Parte b
import numpy as np
import scipy.optimize as opt

def f2(x):
  return 2*x + 3*np.cos(x) - np.exp(x)

def f2prime(x):
  return 2 - 3*np.sin(x) - np.exp(x)

print("Usando método de Newton:", opt.newton(f2,x0=1.5,fprime=f2prime,tol = 10e-5))
print("Usando método Secante:", opt.newton(f2,x0=1,x1=2,tol = 10e-5))
```
